### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



# PASO 0: Unir las cuatro tiendas en un solo DataFrame
# Usamos pd.concat() para juntar los datos uno debajo del otro

```python
datos_completos = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
datos_completos.head()
```

# PASO 1: Análisis de facturación
# Sumamos todos los valores de la columna 'Precio'

```python
facturacion_total = datos_completos['Precio'].sum()
print(f"La facturación total es de ${facturacion_total}")
```

# Mostramos el resultado formateado para que sea fácil de leer

```python
print(f"La facturación total de todas las tiendas es: ${facturacion_total:,.2f}")
```

# 2. Ventas por categoría

# PASO 2: Ventas por categoría
# Usamos groupby para agrupar por la columna 'Categoría del Producto'
# Luego seleccionamos la columna 'Precio' y sumamos sus valores

```python
ventas_por_categoria = datos_completos.groupby('Categoría del Producto')['Precio'].sum()
print(ventas_por_categoria)
```

# Ordenamos los resultados de mayor a menor (descendente)

```python
ventas_por_categoria = ventas_por_categoria.sort_values(ascending=False)
print(ventas_por_categoria)
```

# Mostramos el resultado en pantalla

```python
print("Ventas totales por categoría:")
print(ventas_por_categoria)
```

# 3. Calificación promedio de la tienda


# PASO 3: Calificación promedio de la tienda
# Seleccionamos la columna 'Calificación' y usamos .mean() para obtener el promedio

```python
calificacion_promedio = datos_completos['Calificación'].mean()
print(f"La calificación promedio de la tienda es: {calificacion_promedio}")
```

# Mostramos el resultado limitando a 2 decimales para que sea más claro

```python
print(f"La calificación promedio de toda la cadena de tiendas es: {calificacion_promedio:.2f} estrellas sobre 5")
```

# 4. Productos más y menos vendidos

# PASO 4: Productos más y menos vendidos
# Contamos cuántas veces aparece cada producto en la columna 'Producto'

```python
ventas_por_producto = datos_completos['Producto'].value_counts()
print(ventas_por_producto)
```

# Obtenemos los 5 primeros (más vendidos) y los 5 últimos (menos vendidos)

```python
mas_vendidos = ventas_por_producto.head(5)
menos_vendidos = ventas_por_producto.tail(5)
```

# Mostramos los resultados en pantalla

```python
print(" TOP 5 PRODUCTOS MÁS VENDIDOS:")
print(mas_vendidos)
print("-" * 30) # Esto solo imprime una línea separadora
print("TOP 5 PRODUCTOS MENOS VENDIDOS:")
print(menos_vendidos)
```

# 5. Envío promedio por tienda

# PASO 5: Envío promedio por tienda (Lugar de Compra)
# Agrupamos por ciudad y calculamos el promedio del costo de envío

```python
envio_promedio_ciudad = datos_completos.groupby('Lugar de Compra')['Costo de envío'].mean()
print(envio_promedio_ciudad)
```

# Ordenamos los resultados de mayor a menor costo

```python
envio_promedio_ciudad = envio_promedio_ciudad.sort_values(ascending=False)
print(envio_promedio_ciudad)
```

# Mostramos el resultado en pantalla

```python
print("Costo de envío promedio por tienda/ciudad:")
print(envio_promedio_ciudad)
```
