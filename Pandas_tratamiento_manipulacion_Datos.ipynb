```python
import pandas as pd
```

```python
datos = pd.read_json('/content/datos_hosting.json')
```

```python
datos.head()
```

```python
datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()
```

```python
colmunas = list(datos.columns)
colmunas
```

```python
datos = datos.explode(colmunas[3:])
datos
```

```python
datos.reset_index(drop=True, inplace=True)
datos.head()
```

```python
datos.info()
```

```python
import numpy as np
```

```python
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
```

```python
datos.info()
```

```python
col_numericas = ['cantidad_baños','cantidad_cuartos','cantidad_camas']
```

```python
datos[col_numericas] = datos[col_numericas].astype(np.int64)
```

```python
datos.info()
```

```python
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
datos.info()
```

```python
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
datos.info()
```

```python
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

```python
datos[['cuota_deposito','cuota_limpieza']] = datos[['cuota_deposito','cuota_limpieza']].applymap(lambda x: x.replace('$','').replace(',',''))
```

```python
datos[['cuota_deposito','cuota_limpieza']] = datos[['cuota_deposito','cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

```python
datos.head()
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
datos.head()
```

```python
datos['descripcion_local'][3169]
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace(r"[^a-zA-Z0-9\-\']", " ", regex=True)
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace(
    r'(?<!\w)-(?!\w)', ' ', regex=True
)
```

```python
datos['descripcion_local'].head()
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.split()
datos.head()
```

```python
datos['comodidades'].str.replace('\{|}|\"','', regex=True)
```

```python
datos['comodidades'] = datos['comodidades'].str.split(',')
datos.head
```
